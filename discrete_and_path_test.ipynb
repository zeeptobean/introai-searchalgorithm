#### TSP with 5 cites
rng seed = 113

```python
distance_matrix=np.array([
    [ 0, 48, 65, 68, 68],
    [10,  0, 10, 35, 20],
    [5, 0,  0, 40, 0],
    [65, 0, 0,  0, 80],
    [85, 100, 55, 0, 0]
])

problem = discfunc.TSPFunction(distance_matrix=distance_matrix, dimension=5)

seed = 113

results: list[DiscreteResult] = []
results.append(differential_evolution_discrete(problem, rng_seed=seed))
results.append(genetic_algorithm_discrete(problem, rng_seed=seed))
results.append(simulated_annealing_linear_discrete(problem, rng_seed=seed))
results.append(tlbo_discrete(problem, rng_seed=seed))
results.append(cuckoo_search_discrete(problem, rng_seed=seed))
results.append(pso_discrete(problem, rng_seed=seed))
results.append(abc_discrete(problem, rng_seed=seed))
results.append(firefly_discrete(problem, rng_seed=seed))
results.append(aco_discrete(problem, rng_seed=seed))

plot1 = visualize_convergence_multiple(results, dark_theme=dark_theme)
plot2 = visualize_runtime_vs_best(results, dark_theme=dark_theme)
plot3, _ = visualize_tsp_multiple(results, dark_theme=dark_theme)
plot1.show()
plot2.show()
plot3.show()
```
