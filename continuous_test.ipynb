#### Continuous function test: RastriginFunction(dimension=2), rng_seed = 42

```python
problem = contfunc.RastriginFunction(dimension=2)

rng_seed = 42

results: list[ContinuousResult] = []

results.append(differential_evolution_continuous(problem, rng_seed=rng_seed))
results.append(genetic_algorithm_continuous(problem, rng_seed=rng_seed))
results.append(simulated_annealing_continuous(problem, rng_seed=rng_seed))
results.append(tlbo_continuous(problem, rng_seed=rng_seed))
results.append(cuckoo_search_continuous(problem, rng_seed=rng_seed))
results.append(pso_continuous(problem, rng_seed=rng_seed))
results.append(abc_continuous(problem, rng_seed=rng_seed))
results.append(firefly_continuous(problem, rng_seed=rng_seed))
acoresult = aco_continuous(problem, rng_seed=rng_seed)
results.append(acoresult)

# Visualize the optimization process of ACOR in 2d and 3d
plot1 = visualize_2d(acoresult, dark_theme=dark_theme, export_mp4="video_aco.mp4")    # export 2d animation as mp4 video
plot2 = visualize_3d(acoresult, dark_theme=dark_theme)

# Global comparison
plot3 = visualize_convergence_multiple(results, dark_theme=dark_theme)
plot4 = visualize_runtime_vs_best(results, dark_theme=dark_theme)
plot3.show()
plot4.show()
```

**Retune Simulated Annealing for better convergence**

```python
results[2] = simulated_annealing_continuous(problem, temp=200, cooling_rate=0.99, step_bound=2.5, min_temp=1e-5, rng_seed=rng_seed)

plot3 = visualize_convergence_multiple(results, dark_theme=dark_theme)
plot4 = visualize_runtime_vs_best(results, dark_theme=dark_theme)
plot3.show()
plot4.show()
```
